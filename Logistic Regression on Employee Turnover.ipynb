```python

# In[1]:


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


# In[2]:


train = pd.read_csv("C:/Users/wangyajuan/Desktop/essay/dataset/train.csv")
train = train.drop(['Over18','StandardHours','EmployeeNumber'],axis=1)

cols = list(train)
cols.insert(0,cols.pop(cols.index('Attrition')))
train = train.loc[:,cols]#把Attrtion放到最前面一列
train.shape


# In[3]:


#将离散型特征转化为one-hot编码
cata_result = pd.DataFrame() #初始化生成一个数据框
train_afr = pd.DataFrame()
cata = pd.DataFrame()
cols_to_keep = []
cols_to_alert = []
for i in train.columns: 
    if train[i].dtype == 'int64':
        cols_to_keep.append(i)
for i in train.columns:
    if train[i].dtype == 'O':
        cols_to_alert.append(i)
for i in cols_to_alert:  
    cata = pd.get_dummies(train[i],prefix=i)  
    cata = cata.iloc[:,1:]#iloc行数进行行切片，join合并到原来的变量上
    cata_result = pd.concat([cata_result,cata],axis=1)
train_afr = train[cols_to_keep].join(cata_result)
train = train_afr
train.shape


# In[4]:


X,y = train.iloc[:,1:],train.iloc[:,0]#切片
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)#随机选择25%作为测试集，剩余作为训练集
test = pd.concat([X_test,y_test],axis=1)
logit = sm.Logit(y,X)#statmodel库

lr = LogisticRegression(solver='liblinear',max_iter=1000000)
lr.fit(X_train, y_train)


# 拟合模型
result = logit.fit()
#print(result.summary().as_text)

# In[5]:

y_predict = lr.predict(X_test)
print(y_predict)
print("mes of Linear Regresion squares is", np. mean(( y_predict - y_test ) ** 2))
print("accuracy of Linear regression is",lr.score(X_test,y_test))



#开始预测
#predict_value = result.predict(X_test)
#predict_value.name = 'Predict_Value'#series的命名
#test = pd.concat([test,predict_value],axis=1)
#predict = pd.DataFrame()
#predict.column = 'Predict'
#test = pd.concat([test,predict],axis=1)
#test['Predict']=np.where(test.Predict_Value >= 0.5,1,0)
#m,n = test.shape
#
#hit = 0
#for i in range(m):
#    if test['Predict'].iloc[i] == test['Attrition'].iloc[i]:
#        hit += 1


#输出结果
#print("Hit Rate: %.2f%%" %float(100*hit/m))
#test.to_csv("C:/Users/wangyajuan/Desktop/essay/dataset/testresult.csv")

#lr = LogisticRegression()
#lr.fit(X, y)
#print("得出来的权重：", lr.coef_)

# In[6]:


ceshi = pd.read_csv("C:/Users/wangyajuan/Desktop/essay/dataset/test.csv")
ceshi = ceshi.drop(['Over18','StandardHours','EmployeeNumber'],axis=1)
ceshi['BusinessTravel'].dtype


# In[7]:



#将离散型特征转化为one-hot编码
ceshi_afr = pd.DataFrame()
ceshi_cata = pd.DataFrame()
ceshi_cata_result = pd.DataFrame()
ceshi_cols_to_keep = []
ceshi_cols_to_alert = []
for i in ceshi.columns: 
    if ceshi[i].dtype == 'int64':
        ceshi_cols_to_keep.append(i)
for i in ceshi.columns:
    if ceshi[i].dtype == 'O':
        ceshi_cols_to_alert.append(i)
for i in ceshi_cols_to_alert:  
    ceshi_cata = pd.get_dummies(ceshi[i],prefix=i)  
    ceshi_cata = ceshi_cata.iloc[:,1:]
    ceshi_cata_result = pd.concat([ceshi_cata_result,ceshi_cata],axis=1)
ceshi_afr = ceshi[ceshi_cols_to_keep].join(ceshi_cata_result)
ceshi = ceshi_afr
ceshi.head()


# In[8]:


#开始预测
Z = ceshi.iloc[:,0:]
ceshi_predict_value = result.predict(Z)


# In[9]:


ceshi_predict_value.name = 'Test_Predict_Value'#series的命名
ceshi = pd.concat([ceshi,ceshi_predict_value],axis=1)
ceshi_predict = pd.DataFrame()
ceshi_predict.column = 'Test_Predict'
ceshi = pd.concat([ceshi,ceshi_predict],axis=1)
ceshi['Test_Predict']=np.where(ceshi.Test_Predict_Value >= 0.5,1,0)
ceshi.to_csv("C:/Users/wangyajuan/Desktop/essay/dataset/ceshiresult.csv")

```
